```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

data = pd.read_excel('adp.xlsx')
print(data)
data.columns = ['X', 'Y', 'Place']
data 
cls = LogisticRegression()
scaler = StandardScaler()
x_transform=scaler.fit_transform(data[['X', 'Y']])

cls.fit(x_transform, data[['Place']])
cls.predict(x_transform)

data[['Place']]
c=cls.intercept_
c

m1 = cls.coef_[0, 0]
m2 = cls.coef_[0, 1]

m1, m2
def get_line_y(x):
    return -(c+m1*x)/m2
line_x = np.arange(x_transform[:, 0].min(), x_transform[:,0].max(), 0.1)
plt.plot(line_x,get_line_y(line_x))
plt.scatter(x_transform[:,0],x_transform[:, 1],color='green')
```
