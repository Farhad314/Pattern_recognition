```python
import pandas as pd
import matplotlib.pyplot as plt
# Read the Excel file into a pandas DataFrame
data = pd.read_excel('13.MSE.xlsx')
# Initialize an empty list to store the squared errors
squared_errors = []
# Iterate over each row in the DataFrame
for index, row in data.iterrows():
    # Extract the true value (TV) and predicted value (PV) from the current row
    x = row['x']
    y = row['y']
    
    # Calculate the squared error for the current row and append it to the list
    squared_error = (x - y) ** 2
    squared_errors.append(squared_error)
# Add a new column 'Squared Error' to the DataFrame with the squared error values
data['Squared Error'] = squared_errors
# Save the updated DataFrame to a new Excel file with the squared error column
data.to_excel('updated_data_file.xlsx', index=False)
# Print the updated DataFrame with the squared error column
print(data)
# Plot the scatter plot
plt.scatter(data['x'], data['y'], c='blue', label='Data Points')
# Plot the decision boundary at MSE value
mse = sum(squared_errors) / len(squared_errors)
print("Mean Squared Error (MSE):", mse)
plt.axhline(y=mse, color='red', linestyle='--', label=f'MSE Boundary: {mse:.2f}')
# Add labels and legend
plt.xlabel('TV')
plt.ylabel('PV')
plt.legend()
# Show the plot
plt.show()
```
